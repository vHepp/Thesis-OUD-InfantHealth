```python
import pyreadstat


# formats_as_category is by default True, and it means the replaced values will be transformed to a pandas category column. There is also formats_as_ordered_category to get an ordered category, this by default is False.
df1, meta = pyreadstat.read_sas7bdat("./PRAMS ARF/sas/phase8_arf_2016_2021.sas7bdat",
                                     formats_as_category=True, formats_as_ordered_category=False)
df2, meta = pyreadstat.read_sas7bdat("./PRAMS ARF/sas/phase8_2016_2021_std_drug_miha.sas7bdat",
                                     formats_as_category=True, formats_as_ordered_category=False)
df3, meta = pyreadstat.read_sas7bdat("./PRAMS ARF/sas/phase8_2016_2021_std_u.sas7bdat",
                                     formats_as_category=True, formats_as_ordered_category=False)
```

```python
merged_df = pd.merge(pd.merge(df1, df2, on="ID"), df3,
                     on="ID").sort_values(by="ID")
merged_df
```

```python
df1.to_csv("./PRAMS ARF/csv/phase8_arf_2016_2021.csv", index=False)
df2.to_csv("./PRAMS ARF/csv/phase8_2016_2021_std_drug_miha.csv", index=False)
df3.to_csv("./PRAMS ARF/csv/phase8_2016_2021_std_u.csv", index=False)
merged_df.to_csv("./PRAMS ARF/csv/phase8_2016_2021_merged.csv", index=False)
```
