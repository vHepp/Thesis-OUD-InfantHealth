# Insurance stats, core questionare base


Major caveat: it's impossible to tell why some questions were left blank. Some speculation is necessary (e.g. does blank mean no?)


```python
import pandas as pd
```

```python
arf_df = pd.read_csv("./PRAMS ARF/csv/phase8_arf_2016_2021.csv")
op_blwthr_df = pd.read_csv("./Opioid Supp/phase8_sup_opioid_blwthr.csv")


iso_abusers = pd.read_csv("./isolated_abusers.csv")

print(arf_df.shape)
print(iso_abusers.shape)
```

```python
arf_iso = pd.concat([arf_df, op_blwthr_df])
arf_iso = arf_iso[arf_iso["ID"].isin(iso_abusers["ID"])]

backup = arf_iso.copy()

print(f"{arf_iso.shape[0]} abusers")
```

```python
cols_9s = [
    "INSWORK8",
    "INSPAR",
    "INSHCEX",
    "INSMED",
    "INSCHIP",
    "INSGOV",
    "INSGOV2",
    "INSMIL",
    "INSIHS",
    "INSOTH"
]

cols_10s = [
    "PP8_WORK_RAW",
    "PP8_PAR_RAW",
    "PP8_HCEX_RAW",
    "PP_MEDIC_RAW",
    "PP_CHIP_RAW",
    "PP_GOV_RAW",
    "PP_GOV2_RAW",
    "PP_MILIT_RAW",
    "PP_IHS_RAW",
    "PP_OTH_RAW"
]

cols_11s = [
    "HI_WORK8",
    "HI_PAR",
    "HI_HCEX",
    "HI_MEDIC",
    "HI_CHIP",
    "HI_GOV",
    "HI_GOV2",
    "HI_MILIT",
    "HI_IHS",
    "HI_OTH"
]

cols_all = cols_9s + cols_10s + cols_11s
```

```python
cols_never = [
    "INSNONE",
    "PP_NONE_RAW",
    "HI_NONE"
]
```

```python
cols_TF = [
    "Before",
    "During",
    "After"
]

arf_iso[cols_TF] = False
```

## Skipped insurance section


```python
df_blank = arf_iso[arf_iso[cols_all + cols_never].isna().all(axis=1)].copy()
df_blank[["ID"] + cols_all + cols_never]
```

```python
print(f"{df_blank.shape[0]} women skipped the insurance section")
```

Remove skipped from dataset


```python
print(arf_iso.shape)
arf_iso = arf_iso.drop(df_blank.index)
print(arf_iso.shape)
```

```python
print(f"{arf_iso.shape[0]} women answered parts the insurance section")
```

## Insured Before pregnancy


```python
df_before = arf_iso[(
    arf_iso[cols_9s].eq(2).any(axis=1)
)].copy()

df_before["Before"] = True

# df_before[["ID"] + cols_9s + cols_TF]
```

```python
print(f"{df_before.shape[0]} women insured before pregnancy")
```

```python
arf_iso.loc[df_before.index, "Before"] = True

arf_iso[["ID"] + cols_9s + cols_TF]
```

```python
arf_iso[
    arf_iso["Before"].eq(True)
].shape[0]
```

## Insured During pregnancy


```python
df_during = arf_iso[(
    arf_iso[cols_10s].eq(2).any(axis=1)
)].copy()

df_during["During"] = True

# df_during[["ID"] + cols_10s + cols_TF]
```

```python
print(f"{df_during.shape[0]} women insured during pregnancy")
```

```python
arf_iso.loc[df_during.index, "During"] = True

arf_iso[["ID"] + cols_10s + cols_TF]
```

```python
arf_iso[
    arf_iso["During"].eq(True)
].shape[0]
```

## Insured After pregnancy


```python
df_after = arf_iso[(
    arf_iso[cols_11s].eq(2).any(axis=1)
)].copy()

df_after["After"] = True

df_after[["ID"] + cols_11s + cols_TF]
```

```python
print(f"{df_after.shape[0]} women insured after pregnancy")
```

```python
arf_iso["After"] = False
# arf_iso["Before"].loc[df_before.index] = True
arf_iso.loc[df_after.index, "After"] = True
arf_iso[["ID"] + cols_11s + cols_TF]
```

```python
arf_iso[
    arf_iso["After"].eq(True)
].shape[0]
```

## NOT insured during pregnancy


```python
df_not_during = arf_iso[(
    arf_iso[cols_10s].ne(2).all(axis=1)  # 1's or blanks in 10s
)].copy()

df_not_during[["ID"] + cols_10s]
```

```python
print(f"{df_not_during.shape[0]} women NOT insured during pregnancy")
```

```python
arf_iso[
    arf_iso["During"].eq(False)
].shape[0]
```

## During only


```python
df_during_only = arf_iso[(
    arf_iso[cols_9s].ne(2).all(axis=1) &
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].ne(2).all(axis=1)
)].copy()


df_during_only[["ID"] + cols_9s + cols_10s + cols_11s]
```

```python
print(f"{df_during_only.shape[0]} women only insured during pregnancy")
```

```python
arf_iso[
    arf_iso["Before"].eq(False) &
    arf_iso["During"].eq(True) &
    arf_iso["After"].eq(False)
].shape[0]
```

## During and After, not before


```python
df_during_after_not_before = arf_iso[(
    arf_iso[cols_9s].ne(2).all(axis=1) &
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].eq(2).any(axis=1)
)].copy()


df_during_after_not_before[["ID"] + cols_9s + cols_10s + cols_11s]
```

```python
print(f"{df_during_after_not_before.shape[0]} women only insured during and after pregnancy")
```

```python
arf_iso[
    arf_iso["Before"].eq(False) &
    arf_iso["During"].eq(True) &
    arf_iso["After"].eq(True)
].shape[0]
```

## Insured before and during, not after


```python
df_before_during_not_after = arf_iso[(
    arf_iso[cols_9s].eq(2).any(axis=1) &
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].ne(2).all(axis=1)
)].copy()


df_before_during_not_after[["ID"] + cols_9s + cols_10s + cols_11s]
```

```python
print(f"{df_before_during_not_after.shape[0]} women only insured before and during pregnancy")
```

```python
arf_iso[
    arf_iso["Before"].eq(True) &
    arf_iso["During"].eq(True) &
    arf_iso["After"].eq(False)
].shape[0]
```

## Insured during and after


```python
df_during_after = arf_iso[(
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].eq(2).any(axis=1)
)].copy()


df_during_after[["ID"] + cols_9s + cols_10s + cols_11s]
```

```python
print(f"{df_during_after.shape[0]} women insured during and after pregnancy")
```

```python
arf_iso[
    arf_iso["During"].eq(True) &
    arf_iso["After"].eq(True)
].shape[0]
```

## Lost insurance after pregnancy


```python
df_lost = arf_iso[(
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].ne(2).all(axis=1)
)].copy()


df_lost[["ID"] + cols_9s + cols_10s + cols_11s]
```

```python
arf_iso[
    arf_iso["During"].eq(True) &
    arf_iso["After"].eq(False)
].shape[0]
```

```python
df_all_points = arf_iso[(
    arf_iso[cols_9s].eq(2).any(axis=1) &
    arf_iso[cols_10s].eq(2).any(axis=1) &
    arf_iso[cols_11s].eq(2).any(axis=1)
)].copy()


df_all_points[["ID"] + cols_9s + cols_10s + cols_11s]
```

## Never insured


```python
df_never = arf_iso[(
    arf_iso[cols_never].eq(2).all(axis=1)  # &
    # arf_iso[cols_9s].ne(2).all(axis=1) &
    # arf_iso[cols_10s].ne(2).all(axis=1) &
    # arf_iso[cols_11s].ne(2).all(axis=1)
)].copy()


df_never[["ID"] + cols_never]
```

# Save groups to csv


```python
import os

if not os.path.exists("./Cohorts"):
    os.makedirs("./Cohorts")

df_any_point.to_csv("./Cohorts/any_points.csv", index=False)
df_all_points.to_csv("./Cohorts/all_points.csv", index=False)

df_before.to_csv("./Cohorts/before.csv", index=False)  # newest
df_during.to_csv("./Cohorts/during.csv", index=False)  # newer
df_after.to_csv("./Cohorts/after.csv", index=False)  # newest

df_not_during.to_csv("./Cohorts/not_during.csv", index=False)  # newer
df_during_only.to_csv("./Cohorts/during_only.csv", index=False)

df_during_after_not_before.to_csv("./Cohorts/during_after_not_before.csv", index=False)
df_before_during_not_after.to_csv("./Cohorts/before_during_not_after.csv", index=False)
df_during_after.to_csv("./Cohorts/during_after.csv", index=False)

df_lost.to_csv("./Cohorts/lost.csv", index=False)
df_never.to_csv("./Cohorts/never.csv", index=False)
```
