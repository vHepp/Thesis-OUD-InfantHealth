```python
import pandas as pd
import matplotlib as plt
```

```python
arf_df = pd.read_csv("./PRAMS ARF/csv/phase8_arf_2016_2021.csv")
std_u_df = pd.read_csv("./PRAMS ARF/csv/phase8_2016_2021_std_u.csv")
miha_df = pd.read_csv("./PRAMS ARF/csv/phase8_2016_2021_std_drug_miha.csv")
op_df = pd.read_csv("./Opioid Supp/phase8_sup_opioid.csv")
op_blwthr_df = pd.read_csv("./Opioid Supp/phase8_sup_opioid_blwthr.csv")

print(arf_df.shape)
print(std_u_df.shape)
print(miha_df.shape)
print(op_df.shape)
print(op_blwthr_df.shape)
```

```python
def overlap(df, cols, thing):
    for i in range(1, len(cols)-1):
        for j in range(i+1, len(cols)):
            # print(f"{i}: {cols[i]} vs {j}: {cols[j]}")
            print(
                f'{cols[i]} and {cols[j]}: {df[(df[cols[i]] == thing) & (df[cols[j]] == thing)].shape[0]}\n')
```

```python
arf_df.head()
```

## From phase8_2016_2021_std_u


```python
df1 = pd.merge(arf_df, std_u_df, on="ID", how="inner")
```

```python
df1.shape
```

```python
df1.columns
```

```python
cols = [
    "ID",
    "RXPN_OLDRX",
    "RXPN_NORX",
    "PNC_DRGTRT"
]

df_std_u = df1[cols].copy().dropna(subset=cols[1:], how='all')

df_std_u
```

```python
df_std_u = df_std_u[df_std_u[cols].eq(2).any(axis=1)]
df_std_u
```

```python
overlap(df_std_u, cols, thing=2)
```

## From phase8_2016_2021_std_drug_miha


```python
df2 = pd.merge(arf_df, miha_df, on="ID", how="inner")
```

```python
df2.shape
```

```python
cols_miha = [
    "ID",
    "DRGB_MDN",
    "DRGB_HRN",
    "NJDR_MDN",
    "NJDR_HRN",
    "DRGP_MDN",
    "DRGP_HRN"
]

df_miha = df2[cols_miha].copy().dropna(subset=cols_miha[1:], how='all')
df_miha
```

```python
df_miha = df_miha[df_miha[cols_miha].eq(2).any(axis=1)].drop_duplicates()
df_miha
```

```python
for col in (cols_miha[1:]):
    counts = df_miha[col].value_counts()
    non = df_miha[col].isna().sum()

    print(f"{counts}\nBlank\t{non}\n")
```

```python
overlap(df_miha, cols_miha, thing=2)
```

## Save isolated IDs


```python
std_ids = df_std_u[["ID"]]
miha_ids = df_miha[["ID"]]
op_ids = df_op[["ID"]]
blw_ids = df_blwthr[["ID"]]
```
